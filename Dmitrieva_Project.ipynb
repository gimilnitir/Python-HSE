```python
#As we can see, the nltk lemmatization is not as good as pymorphy one. 
#The analytical verb forms are interpreted as several words, and their lemmas are not reconstructed.
#Even more confusing is the case of "was": it is analyzed not as past of "to be", but as a plural for "*wa"...
#However, I do not concentrate on morphological features here.
#There is an advantage: the letters are not lowered automatically, so the name Brown will not be mistaken for the adjective "brown". 

entokens = nltk.word_tokenize(entext)
print(entokens)
```

```python
#doing POS tagging with nltk

lemmatizer = WordNetLemmatizer()
pos_tagged = nltk.pos_tag(entokens)  
print(pos_tagged)
```

```python
#The POS tags in NLTK are too complicated and not always correct. We create a new function to do more general POS marking.

def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:         
        return None
```

```python
wordnet_tagged = list(map(lambda x: (x[0], pos_tagger(x[1])), pos_tagged))
print(wordnet_tagged)
```

```python
#Now I count the percentage of different parts of speech in Russian and English

runoun=0
ruverb=0
ruadj=0
ruadv=0
for x in range(len(analysis)):
    if "NOUN" in analysis[x].values():
        runoun = runoun+1
    if "VERB" in analysis[x].values():
        ruverb = ruverb+1
    if "ADJS" in analysis[x].values():
        ruadj = ruadj+1
    if "ADJF" in analysis[x].values():
        ruadj = ruadj+1 
    if "ADVB" in analysis[x].values():
        ruadv = ruadv+1
```

```python
ennoun=0
enadj=0
enadv=0
enverb=0
for word, tag in wordnet_tagged:
    if tag is "n":
        ennoun= ennoun+1
    if tag is "a":
        enadj= enadj+1
    if tag is "r":
        enadv= enadv+1
    if tag is "v":
        enverb= enverb+1
```

```python
finaldata = [
    {"pos": "noun", "Russian": runoun/len(rutokens), "English": ennoun/len(entokens)},
    {"pos": "verb", "Russian": ruverb/len(rutokens), "English": enverb/len(entokens)}, 
    {"pos": "adjective", "Russian": ruadj/len(rutokens), "English": enadj/len(entokens)}, 
    {"pos": "adverb", "Russian": ruadv/len(rutokens), "English": enadv/len(entokens)}
]

df = pd.DataFrame(finaldata)
df
```
